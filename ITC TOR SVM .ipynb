# Normalization (Maximum absolute Scaling )

https://www.geeksforgeeks.org/normalize-a-column-in-pandas/

```python
#normalization 
df_max_scaled = df.copy()
  
# apply normalization techniques
for column in df_max_scaled.columns:
    df_max_scaled[column] = df_max_scaled[column]  / df_max_scaled[column].abs().max()
      
# view normalized data
df_max_scaled.head()
```

## Applying Linear SVC  SVM  classification model 

```python
# creating subset of features  
# Removing sourceIP,'Source IP',source port destination IP destination port protocol 
             

features_all = [c for c in df.columns if c not in ['label']]

features = [c for c in df.columns if c not in
            ['SourceIP',
             'SourcePort',
             'DestinationIP',
             'DestinationPort',
             'Protocol',
             'label'
             
             ]]
features
```

```python
df["label"].value_counts()                                         #  imbalance dataset 
```

```python
X = np.array(df[features])
X_all = np.array(df[features_all])
y = np.array(df.label)
print(X.shape, np.mean(y))
```

```python

from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test = train_test_split(X,y, test_size = 0.20)

#print(cl('Train set shape :', attrs = ['bold']), X_train.shape, y_train.shape)
#print(cl('Test set shape :', attrs = ['bold']), X_test.shape, y_test.shape)
```

https://medium.com/analytics-vidhya/introduction-to-svm-and-kernel-trick-part-1-theory-d990e2872ace

```python
from sklearn.svm import SVC 
svclassifer = SVC(kernel = 'linear')
svclassifer.fit(X_train, y_train)
```

```python
y_pred = svclassifer.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test,y_pred), "\n")
"""
           |                  Actual               
           |-----------------------------------------
           |  True  Positive   |   False Positive   |
           |                   |                    |
Predicted  |-----------------------------------------
           |                   |   True Negative    |
           |  False Negative   |                    |
           |-----------------------------------------
"""



print(classification_report(y_test,y_pred))
```
